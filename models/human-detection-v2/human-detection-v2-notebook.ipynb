```python
import os
import cv2
import torch

# ng dn n file m hnh
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), '..'))
model_path = os.path.join(PROJECT_ROOT, 'human-detection-v2/best.pt')

# Ti m hnh
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path)

# Khi ng webcam
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # D on
    results = model(frame)

    # Hin th kt qu
    results.render()  # V bounding boxes ln frame
    cv2.imshow('Webcam', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Gii phng ti nguyn
cap.release()
cv2.destroyAllWindows()
```
